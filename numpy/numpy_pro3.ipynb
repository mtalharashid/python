## Advance Operation with Business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data structure [Restaurant_id, 2022, 2023, 2024, 2025]

sales_data = np.array([
    [1, 16000, 18000, 22000, 41000],
    [2, 21000, 31000, 20000, 5000],
    [3, 21000, 31000, 20000, 15000],
    [4, 11000, 21000, 17000, 20000],
    [5, 21000, 26000, 27000, 30000],
])

print("--- zomato sales analysis ---", )
print("\n sales data shape", sales_data.shape)
print("\n sample data for first 3 Restaurant", sales_data[0:3]) # start from zero index till 2nd index 0,1,2 
print("\n sample data for first column", sales_data[:,1:]) # start from 1st column index 
```

```python
# Find the total sale of each restaurant 

print(np.sum(sales_data[:, 1:], axis=0)) # skip the 0 index of column and start from 1 

yearly_total = np.sum(sales_data[:, 1:], axis=0) # " : " represent all rows after ,  " 1: " start from column 1

print("yearly Data of sale: ", yearly_total)

print("sum of all years: ", sum(yearly_total), "$")

# minimum Sale per restaurant 
minimum_sale = np.min(sales_data[:, 1:], axis=1)
print("Minimum Sale of a Restaurant:", minimum_sale)

# maximum Sale per restaurant 
maximum_sale = np.max(sales_data[:, 1:], axis=1)
print("Maximum Sale of a Restaurant:", maximum_sale)

# maximum Sale per year 
maximum_sale_year = np.max(sales_data[:, 1:], axis=0)
print("Maximum Sale per year:", maximum_sale_year)

#Average sale per restaurant 
avrg_sale = np.mean(sales_data[:, 1:], axis=1)
print("Average sales per restaurant: ", avrg_sale)

#find Cumulative sales 
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sum sale of restaurants: \n", cumsum)

plt.figure(figsize=(10, 5))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Avarage Cumulative sales per year across all restaurant: ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
